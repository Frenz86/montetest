## DISTRIBUZIONE UNIFORME


```python
import numpy as np
import matplotlib.pyplot as plt

def mc_uniform(min, max,samples):
    "Simulate a Uniform Distribution"
    data = []
    for i in range(samples):
        data.append(np.random.uniform(min, max))
    return np.array(data)
```

```python
# DISTRIBUZIONE UNIFORME

from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def mc_uniform_hist(min, max,samples,bins,title='title',color="green"):
    "Simulate a Uniform Distribution"
    data = []
    for i in range(samples):
        data.append(np.random.uniform(min, max))
    data = np.array(data)
    fig = plt.figure(figsize=(15, 8))
    ax = plt.axes()
    plt.ylabel("Proportion")
    values, base, _ = plt.hist( data, bins = bins, alpha = 0.5, color = color,label = "Simulation")
    ax_bis = ax.twinx()
    values = np.append(values,0)
    ax_bis.plot( base, np.cumsum(values)/ np.cumsum(values)[-1], color='darkorange', marker='o', linestyle='-', markersize = 1, label = "Cumulative Probability" )
    plt.xlabel('labels')
    plt.ylabel("Probability")

    plt.axvline(data.mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.1, max_ylim*0.01, 'Mean: {:.2f}'.format(data.mean()))

    plt.title(title)
    ax_bis.legend();
    ax.legend();
    plt.show()
    return data

interact(mc_uniform_hist,
        min=(500,600,1),
        max=(1000,1200,1),
        samples=(5000,6000,1),
        bins=(200,300,1),
         );
```

## DISTRIBUZIONE NORMALE

```python
import numpy as np
import matplotlib.pyplot as plt

def mc_normal(mean, std_dev,samples):
    "Simulate a Normal Distribution"
    data = []
    for i in range(samples):
        data.append(np.random.normal(mean, std_dev))
    return np.array(data)
```

```python
# DISTRIBUZIONE NORMALE

from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def mc_normal_hist(mean, std_dev,samples,bins,title='title',color="green"):
    "Simulate a Normal Distribution"
    data = []
    for i in range(samples):
        data.append(np.random.normal(mean, std_dev))
    data = np.array(data)
    fig = plt.figure(figsize=(15, 8))
    ax = plt.axes()
    plt.ylabel("Proportion")
    values, base, _ = plt.hist( data, bins = bins, alpha = 0.5, color = color,label = "Simulation")
    ax_bis = ax.twinx()
    values = np.append(values,0)
    ax_bis.plot( base, np.cumsum(values)/ np.cumsum(values)[-1], color='darkorange', marker='o', linestyle='-', markersize = 1, label = "Cumulative Probability" )
    plt.xlabel('labels')
    plt.ylabel("Probability")

    plt.axvline(data.mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.1, max_ylim*0.01, 'Mean: {:.2f}'.format(data.mean()))

    plt.title(title)
    ax_bis.legend();
    ax.legend();
    plt.show()
    return data

interact(mc_normal_hist,
        mean=(500,600,1),
        std_dev=(50,80,1),
        samples=(5000,6000,1),
        bins=(200,300,1),
         );
```

## Check Probabilita'

```python
def plot_hist(data,title='title',color='green',bins=200):
    data = np.array(data)
    fig = plt.figure(figsize=(15, 8))
    ax = plt.axes()
    plt.ylabel("Proportion")
    values, base, _ = plt.hist( data, bins = bins, alpha = 0.5, color = color,label = "Simulation")
    ax_bis = ax.twinx()
    values = np.append(values,0)
    ax_bis.plot( base, np.cumsum(values)/ np.cumsum(values)[-1], color='darkorange', marker='o', linestyle='-', markersize = 1, label = "Cumulative Probability" )
    plt.xlabel('labels')
    plt.ylabel("Probability")

    plt.axvline(data.mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.1, max_ylim*0.01, 'Mean: {:.2f}'.format(data.mean()))

    plt.title(title)
    ax_bis.legend();
    ax.legend();
    plt.show()

# success conditions
def check_prob(total,threshold):
    probability = np.sum(total <= threshold)/len(total)*100
    return print(f"Probability to be below: {threshold} is", round(probability, 2), "%")

from ipywidgets import interactive
def test(total):
    threshold = total.mean()
    return interactive(lambda threshold: (check_prob(total,threshold), threshold), threshold=(int(total.min()),int(total.max()),1))
```

# Deli case

_INPUT_

**Cost of capital: 15%**

**Investment: 120,000 €**

**Average profile tickets**
* Coffees: 50 units, 1.00 € price, 82% contribution margin
* Breakfasts: 20 units, 2.20 € price, 80% contribution margin
* Snacks: 12 units, 3.50 € price, 65% contribution margin
* Beverages: 18 units, 2.00 € price, 70% contribution margin
* Total: 100 units, 1.72 € average price, 75% average contribution margin

**Medium shopping centre**
Average daily tickets: 494 (min 407, max 581)
Fixed costs structure: 286,000 €

**Big shopping centre**
Average daily tickets: 1017 (min 872, max 1163)
Fixed costs structure: 374,000 €


```python
fixedCosts = 374000
costOfCapital = 0.15
investment = 120000

coffeePrice = 1
coffeeCM = 0.82
breakfastPrice = 2.2
breakfastCM = 0.80
snackPrice = 3.5
snackCM = 0.65
beveragePrice = 2
beverageCM = 0.70
```

```python
########################## COFFEES ##########################

coffees_ = interactive(mc_normal_hist,
                        mean=(500,600,1),
                        std_dev=(50,80,1),
                        samples=(4000,6000,1),
                        bins=(200,300,1),
                        title='coffees_sim',
                        color='green',
                        );
display(coffees_)

##########################  BREACKFAST  #####################

breakfasts_ = interactive(mc_normal_hist,
                        mean=(200,400,1),
                        std_dev=(25,50,1),
                        samples=(4000,6000,1),
                        bins=(200,300,1),
                        title='breakfasts_sim',
                        color='green',
                        );
display(breakfasts_)

###########################  SNACKS  ########################

snacks_ = interactive(mc_normal_hist,
                        mean=(100,180,1),
                        std_dev=(18,40,1),
                        samples=(4000,6000,1),
                        bins=(200,300,1),
                        title='snacks_sim',
                        color='green',
                        );
display(snacks_)

###########################  BEVERAGES  ########################

beverages_ = interactive(mc_normal_hist,
                        mean=(100,180,1),
                        std_dev=(18,40,1),
                        samples=(4000,6000,1),
                        bins=(200,300,1),
                        title='beverages_sim',
                        color='green',
                        );
display(beverages_)
```

```python
coffees = coffees_.result
breakfasts = breakfasts_.result
snacks = snacks_.result
beverages = beverages_.result
################################################
dailyTickets = coffees + breakfasts + snacks + beverages
dailyRevenue = coffees * coffeePrice + breakfasts * breakfastPrice + snacks * snackPrice + beverages * beveragePrice
dailyIncome  = coffees * coffeePrice * coffeeCM + breakfasts * breakfastPrice * breakfastCM + snacks * snackPrice * snackCM + beverages * beveragePrice * beverageCM

plot_hist(dailyTickets,'Daily Tickets','blue')
test(dailyTickets)

plot_hist(dailyRevenue,'Daily Revenue','blue')
test(dailyRevenue)

plot_hist(dailyIncome,'Daily Income','blue')
test(dailyIncome)
```

```python
days = 310

yearlyTickets = dailyTickets * days
yearlyRevenues = dailyRevenue * days
yearlyOperatingIncome = dailyIncome * days


plot_hist(yearlyTickets,'Yearly Tickets','red')
test(dailyTickets)

plot_hist(dailyRevenue,'Yearly Revenue','red')
test(dailyRevenue)

plot_hist(dailyIncome,'Yearly Income','red')
test(dailyIncome)
```
