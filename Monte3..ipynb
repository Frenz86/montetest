## DISTRIBUZIONE UNIFORME


```python
import numpy as np
import matplotlib.pyplot as plt

def mc_uniform(min, max,samples):
    "Simulate a Uniform Distribution"
    data = []
    for i in range(samples):
        data.append(np.random.uniform(min, max))
    return np.array(data)
```

```python
# DISTRIBUZIONE UNIFORME

from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def mc_uniform_hist(min, max,samples,bins,title='title',color="green"):
    "Simulate a Uniform Distribution"
    data = []
    for i in range(samples):
        data.append(np.random.uniform(min, max))
    data = np.array(data)
    fig = plt.figure(figsize=(15, 8))
    ax = plt.axes()
    plt.ylabel("Proportion")
    values, base, _ = plt.hist( data, bins = bins, alpha = 0.5, color = color,label = "Simulation")
    ax_bis = ax.twinx()
    values = np.append(values,0)
    ax_bis.plot( base, np.cumsum(values)/ np.cumsum(values)[-1], color='darkorange', marker='o', linestyle='-', markersize = 1, label = "Cumulative Probability" )
    plt.xlabel('labels')
    plt.ylabel("Probability")

    plt.axvline(data.mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.1, max_ylim*0.01, 'Mean: {:.2f}'.format(data.mean()))

    plt.title(title)
    ax_bis.legend();
    ax.legend();
    plt.show()
    return data

interact(mc_uniform_hist,
        min=(500,600,1),
        max=(1000,1200,1),
        samples=(5000,6000,1),
        bins=(200,300,1),
         );
```

```python
def plot_hist(data,title='title',color='green',bins=200):
    data = np.array(data)
    fig = plt.figure(figsize=(15, 8))
    ax = plt.axes()
    plt.ylabel("Proportion")
    values, base, _ = plt.hist( data, bins = bins, alpha = 0.5, color = color,label = "Simulation")
    ax_bis = ax.twinx()
    values = np.append(values,0)
    ax_bis.plot( base, np.cumsum(values)/ np.cumsum(values)[-1], color='darkorange', marker='o', linestyle='-', markersize = 1, label = "Cumulative Probability" )
    plt.xlabel('labels')
    plt.ylabel("Probability")

    plt.axvline(data.mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.1, max_ylim*0.01, 'Mean: {:.2f}'.format(data.mean()))

    plt.title(title)
    ax_bis.legend();
    ax.legend();
    plt.show()

# success conditions
def check_prob(total,threshold):
    probability = np.sum(total <= threshold)/len(total)*100
    return print(f"Probability to be below: {threshold} is", round(probability, 2), "%")

from ipywidgets import interactive
def test(total):
    threshold = total.mean()
    return interactive(lambda threshold: (check_prob(total,threshold), threshold), threshold=(int(total.min()),int(total.max()),1))
```
